# III - Analyse

### Analyse des valeurs uniques de chaque colonnes

#### Analyse globale

```python
dt_columns = df_train_final.columns.values.tolist()

for i in dt_columns :
    print(f'colonne {i} : {df_train_final[i].unique()}\n')
```

#### Analyse des NaN

```python
incomplet_data = 0
dt_columns = df_train_final.columns.values.tolist()
column_with_nan = []
tab_count_nan = []

for i in dt_columns :
    verif = df_train_final[i].isna().sum()
    if verif != 0 :
        column_with_nan.append(i)
        tab_count_nan.append(df_train_final[i].isna().sum())
        
    incomplet_data += df_train_final[i].isna().sum()

nan_analysis = pd.DataFrame({'Feature' : column_with_nan ,  'Nb_NaN' : tab_count_nan})

print(f"{len(column_with_nan)} colonnes contiennet au moins un 'NaN' :")

nan_analysis
```

```python
dt_columns = df_train_final.columns.values.tolist()
lignes_with_nan = 220600
total_info = len(df_train_final['SK_ID_CURR']) * len(dt_columns)

print(f"Total Data = {total_info}")
print(f"Missing Data = {incomplet_data}")
print(f"Missing Data Rate = {round( (incomplet_data / total_info),2) * 100} %")
print(f"Lignes with NaN = {round( (lignes_with_nan / len(df_train_final['SK_ID_CURR'])),2) * 100} %")
```

### Résultats de l'analyse :

Toutes les lignes contenant des NaN sont à retirer

# III - Nettoyage des données

### Retrait des lignes avec NaN

```python
# on récupère les lignes avec des nan
index_with_nan = df_train_final.index[df_train_final.isnull().any(axis=1)]

# on récupère les id des personnes avec des données manquantes
dossier_incomplet = df_train_final.SK_ID_CURR[df_train_final.isnull().any(axis=1)]
```

```python
df_train_final.drop(index_with_nan,0,inplace=True)
```

```python
df_train_final
```

# IV - Analyse des données après nettoyage

```python
incomplet_data = 0
dt_columns = df_train_final.columns.values.tolist()
column_with_nan = []
tab_count_nan = []

for i in dt_columns :
    verif = df_train_final[i].isna().sum()
    if verif != 0 :
        column_with_nan.append(i)
        tab_count_nan.append(df_train_final[i].isna().sum())
        
    incomplet_data += df_train_final[i].isna().sum()

nan_analysis = pd.DataFrame({'Feature' : column_with_nan ,  'Nb_NaN' : tab_count_nan})

print(f"{len(column_with_nan)} colonnes contiennet au moins un 'NaN' :")

nan_analysis
```

```python
incomplet_data = 0
for i in df_train_final.columns.values.tolist() :      
    incomplet_data += df_train_final[i].isna().sum()


total_info = len(df_train_final['SK_ID_CURR']) * len(dt_columns)

print(f"Total Data = {total_info}")
print(f"Missing Data = {incomplet_data}")
print(f"Missing Data Rate = {round( (incomplet_data / total_info),2) * 100} %")
```
